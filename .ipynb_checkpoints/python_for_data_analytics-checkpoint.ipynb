```python
path = './try_jupyter/pydata-book/ch02/usagov_bitly_data2012-03-16-1331923249.txt'
```

```python
import json
records = [json.loads(line) for line in open(path)]
```

```python
timezones = [rec['tz'] for rec in records if 'tz' in rec]
```

```python
timezones
```

```python
def get_counts(sequence):
    counts = {}
    for x in sequence:
        if x in counts:
            counts[x] += 1
        else:
            counts[x] = 1
    return counts

from collections import defaultdict


def get_counts2(sequence):
    counts = defaultdict(int)
    for x in sequence:
        counts[x] += 1
        
    return counts

counts = get_counts2(timezones)
```
